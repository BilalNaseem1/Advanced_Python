## Variable names

### Single underscore
- If identifier starts with a single underscore, it indicates a private object
- In python nothing is private, so more of internal use
- we can access it but cant import it using statements like *from module import \**

### Double Underscore
- If at start only Is used for name mangling (also called dunder) - changes the variable name in a set way which is useful for inheritence chains
- if dunder at start and end, these are system defined names already defined by python for e.g. \_\_init\_\_

## Functions
- inside () we define parameters
- the function exists as an object, to call it we include paranthesis

```python
def func_1(a: int, b: int): # :int is annotation - what i expect (only documentation)
    return a *b

func_1(2, 3)
```

```python
func_1('a', 3)
```

```python
func_1([1, 2, 3], 3)
```
