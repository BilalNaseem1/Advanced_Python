# Constructors

- A Constructor is the **\_\_init\_\_** function
- It is a special function inside a class, which is invoked at the time of object creation.

- All classes have a function called \_\_init\_\_(), which is always executed when the object is being initiated
- If we dont write an init function, python will automatically create an init function and execute it 

- The init function takes a self parameter.
- Self means the new object which is being created
- The init function can also take other parameters besides self

```python
class Student: # we start name of class with capital letter
    def __init__(self):
        print("adding a new student in the database")
    name = "Bilal"
```

```python
s1 = Student()
```

```python
class Student: # we start name of class with capital letter
    def __init__(self):
        print(self)
        print("adding a new student in the database")
    name = "Bilal"

s1 = Student()
```

- Using the self parameter we can store data - which we call attributes

```python
class Student: # we start name of class with capital letter
    def __init__(self, name, marks):
        self.name = name     # by convention same wording
        self.marks = marks
    college_name = "Oxford University"

s1 = Student("Bilal", 97)
print(s1.name)
print(s1.marks)
print(s1.college_name)
```

There are 2 types of constructors:
1. Default constructors
   - only self argument
   - If we dont create it python will
2. Parameterized constructors


# Attributes

Types of attributes:
1. Class attributes
   - Which is common for all objects - *dont define inside constructor* **(college_name)**
2. Object attributes
   - Different for every object - *define inside constructor* **(self.name)**

- Precedance of object attribute is higher than class attribute

# Methods

Inside a class 2 types of things can be stored:
1. data (attributes)
3. methods

Methods are functions which belong to objects

```python
class Student: # we start name of class with capital letter
    def __init__(self, name, marks):
        self.name = name     # by convention same wording
        self.marks = marks

    def hello(self):
        print("Hello")
```

```python
s1 = Student("Bilal", 97)

s1.name
```

```python
s1.marks
```

```python
s1.hello()
```

- A class will always have a constructor - it is for object initialization/creation

# Static Methods

```python
class Student: # we start name of class with capital letter
    def __init__(self, name, marks):
        self.name = name     # by convention same wording
        self.marks = marks

    def hello():
        print("Hello")

s1 = Student("Bilal", 97)
s1.hello()
```

```python
class Student: # we start name of class with capital letter
    def __init__(self, name, marks):
        self.name = name     # by convention same wording
        self.marks = marks
    @staticmethod
    def hello():
        print("Hello")

s1 = Student("Bilal", 97)
s1.hello()
```
